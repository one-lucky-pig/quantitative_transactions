NunPy提供两个函数polyfit和polyval便利函数来实现基于OLS在简单的单项上，Xx, x2, x3, xn。为了说明起见，请考虑线性、三次和九度OLS回归（见图A-5）：

```python
x = np.arange(len(data.cumsum()))
y = 0.2 * data.cumsum() ** 2
rg1 = np.polyfit(x, y, 1)
rg3 = np.polyfit(x, y, 3)
rg9 = np.polyfit(x, y, 9)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'r', label='data')
plt.plot(x, np.polyval(rg1, x), 'b--', label='linear')
plt.plot(x, np.polyval(rg3, x), 'b-.', label='cubic')
plt.plot(x, np.polyval(rg9, x), 'b:', label='9th degree')
plt.legend(loc=0)
```

## Case Study
当涉及到财务数据时，熊猫软件包中有有用的数据导入功能（另见第3章）。以下代码使用pd. read_csy()函数从存储在远程服务器上的CSV文件中读取$&P 500索引和VIX波动性索引的历史每日数据：

```python
raw = pd.read_csv('http://hilpisch.com/pyalgo_eikon_eod_data.csv',index_col=0, parse_dates=True).dropna()
raw
```

```python
#raw['.SPX']Series对象，pd.DataFrame(raw['.SPX'])是DataFrame对象
spx = pd.DataFrame(raw['.SPX'])
spx.info()
```

```python
vix = pd.DataFrame(raw['.VIX'])
vix.info()
```

```python
spxvix = pd.DataFrame(spx).join(vix)
spxvix.info()
```

```python
#基于左边DataFrame（spx）和右边DataFrame（vix）的索引进行合并，他们的索引都是时间
spxvix = pd.merge(spx, vix,
left_index=True, # merge on left index
right_index=True, # merge on right index
)
spxvix.info()
```

```python
spxvix = pd.DataFrame({'SPX': spx['.SPX'],'VIX': vix['.VIX']},index=spx.index)
spxvix.info()
```

```python
spxvix.plot(figsize=(10, 6), subplots=True)
```

```python
spxvix.plot(figsize=(10, 6))
```

熊猫还允许对整个DataFrame对象进行矢量化操作。以下代码以矢量化的方式同时计算sxxvix对象两列的对数返回。移位方法将数据集按提供的索引值数量移动（在此特定情况下，按一个交易日）：

```python
#spxvix.shift(1) 是将spvxvix DataFrame 中的所有行向下平移一个位置，这样每一行的值将会与前一行的值相比较。这实际上是为了计算收益率，因为股票市场数据中的收益率通常是当前时刻的价格与前一时刻价格的比率。
rets = np.log(spxvix / spxvix.shift(1))#今天/昨天
rets = rets.dropna()
rets.head()
```

```python
rg = np.polyfit(rets['SPX'], rets['VIX'], 1)
rets.plot(kind='scatter', x='SPX', y='VIX',style='.', figsize=(10, 6))
plt.plot(rets['SPX'], np.polyval(rg, rets['SPX']), 'r-')
```

```python
ret = rets.mean() * 252
ret
```

```python
vol = rets.std() * math.sqrt(252)
vol
```

```python
(ret - 0.01) / vol
```

我们只计算标准普尔500指数的最大缩减额更复杂一些。为了计算，我们使用. cummax()方法，该方法记录截至特定日期的时间序列的运行历史最大值。考虑以下在图A-10中生成绘图的代码：

```python
plt.figure(figsize=(10, 6))
spxvix['SPX'].plot(label='S&P 500')
spxvix['SPX'].cummax().plot(label='running maximum')
plt.legend(loc=0)#显示图例的位置，自适应方式
```

The absolute mnaximum drawdown（绝对最大回撤） is the largest difference between the running maximum and the current index level. In our particular case, it is about 580 index points.
The relative maximum drawdown（相对最大回撤） might sometimes be a bit more meaningful. It is here a value of about 20%:

```python
adrawdown = spxvix['SPX'].cummax() - spxvix['SPX']
adrawdown.max()
```

```python
rdrawdown = ((spxvix['SPX'].cummax() - spxvix['SPX']) /spxvix['SPX'].cummax())
rdrawdown.max()
```

The longest drawdown period（最长回撤期） is calculated as follows. The following code selects all those data points where the drawdown is zero (where a new maximum is reached). It then calculates the difference between two consecutive index values (trading dates) for which the drawdown is zero and takes the maximum value. Given the data set we are analyzing, the longest drawdown period is 417 days:

```python
temp = adrawdown[adrawdown == 0]
periods_spx = (temp.index[1:].to_pydatetime() -
                         temp.index[:-1].to_pydatetime())
periods_spx[50:60]
```

```python
periods_spx.max()
```
